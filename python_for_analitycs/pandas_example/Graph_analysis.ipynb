```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
data = pd.read_csv('kc_house_data.csv')
```

```python
data.head()
```

1. Постройте гистограмму, отображающую распределение цен на недвижимость
 в наборе данных из файла kc-house-data.

```python
def format_price(x, pos):
    return f'{int(x):,}'

plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=50, edgecolor='gray')
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_price))
plt.grid(True, which='major', linestyle='-', color='gray')
plt.grid(True, which='minor', linestyle=':', color='gray')
plt.minorticks_on()
plt.show()
```

На графике показано распределение цен на недвижимость. Большинство домов находятся в ценовом диапазоне от 200тыс. до 400тыс. Количество домов дороже 1млн меньше 400 и постепенно снижается.

 3. Постройте гистограмму, показывающую распределение годов постройки
 домов. График должен помочь выявить тенденции в строительстве по годам.

```python
def format_price(x, pos):
    return f'{int(x):,}'

plt.figure(figsize=(10, 6))
plt.hist(data['yr_built'], bins=50, edgecolor='gray')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_price))
plt.grid(True, which='major', linestyle='-', color='gray')
plt.grid(True, which='minor', linestyle=':', color='gray')
plt.minorticks_on()
plt.show()
```

На графике отображено распределение количества домов в соответствии с годом постройки. Прослеживается "омолаживание" домов. Домов до 1945 года на территории мало. Сохраняется своеобразная пиковость некоторых значений. В частности, нельзя не заметить пики около 1975 года и 2008. При этом активное строительство велось примерно каждые 10 лет.

1. Постройте столбчатую диаграмму для анализа количества домов с видом на
 набережную и без вида. График должен показать соотношение между домами с
 видом и без него.

```python
plt.figure(figsize=(6, 6))
data['waterfront'].value_counts().plot(kind='bar')
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'], rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
На графике видно, что подавляющее большинство домов не имеют вида на набережную.
```

 2. Постройте столбчатую диаграмму, отображающую распределение домов по
 количеству этажей. График должен наглядно показать, как часто встречаются
 дома с разным количеством этажей.

```python
plt.figure(figsize=(6, 6))
data['floors'].value_counts().sort_index().plot(kind='bar')
plt.title('Распределение домов по этажности')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

На графике видно, что наиболее распространенными являются 1- и 2-этажные дома.

 1. Постройте диаграмму рассеяния, чтобы изучить зависимость между
 количеством спален и стоимостью недвижимости. График должен показать, как
 изменение количества спален влияет на цену.

```python
def format_price(x, pos):
    return f'{int(x):,}'
plt.figure(figsize=(10, 6))
plt.scatter(data['bedrooms'], data['price'], alpha=0.5)
plt.title('График разброса цены по количеству спален в доме')
plt.xlabel('Количество спален')
plt.ylabel('Цена')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_price))
plt.grid(True, which='major', linestyle='-', color='gray')
plt.grid(True, which='minor', linestyle=':', color='gray')
plt.minorticks_on()
plt.show()
```

 2. Постройте график размаха (boxplot) для анализа влияния состояния дома на
 его стоимость. График должен продемонстрировать, как состояние дома
 связано с его ценой и выявить возможные отклонения.

```python
def format_price(x, pos):
    return f'{int(x):,}'
plt.figure(figsize=(10, 6))
plt.boxplot([data[data['condition'] == i] ['price'] for i in sorted(data['condition'].unique())],
            tick_labels=sorted(data['condition'].unique()))
plt.title('График разброса цены по состоянию домов')
plt.xlabel('Состояние дома')
plt.ylabel('Цена')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_price))
plt.grid(True, which='major', linestyle='-', color='gray')
plt.grid(True, which='minor', linestyle=':', color='gray')
plt.minorticks_on()
plt.show()
```
